# NeuralBridge — EU CRA Compliance Guide

This notebook demonstrates NeuralBridge's built-in compliance features for the **EU Cyber Resilience Act (CRA)**.

**Key deadline: September 11, 2026** — All products with digital elements must comply.

NeuralBridge provides:
1. Immutable audit logging with hash-chain integrity
2. CRA Article 14 vulnerability reporting
3. CycloneDX SBOM generation
4. Structured incident logging
5. GDPR Article 30 processing activity register

```python
from datetime import datetime, timezone

deadline = datetime(2026, 9, 11, tzinfo=timezone.utc)
days_remaining = (deadline - datetime.now(timezone.utc)).days
print(f'Days until CRA deadline: {days_remaining}')
```

## 1. Audit Logging

Every operation through NeuralBridge is logged in an immutable, hash-chain-verified audit trail.

```python
from neuralbridge.security.audit import AuditLogger

audit = AuditLogger()

# Log some events
await audit.log_event(
    event_type='adapter.execute',
    actor='openclaw-agent-1',
    resource='salesforce',
    action='query',
    result='success',
)

events = await audit.query_events(filters={})
print(f'Audit events: {len(events)}')
```

## 2. CRA Vulnerability Report

```python
from neuralbridge.compliance.cra_report import CRAReportGenerator

gen = CRAReportGenerator()
report = gen.generate_vulnerability_report()
print('Vulnerability Report:')
for key, value in report.items():
    print(f'  {key}: {value}')
```

## 3. Software Bill of Materials (SBOM)

```python
from neuralbridge.compliance.sbom import SBOMGenerator

sbom_gen = SBOMGenerator()
sbom = sbom_gen.generate()
print(f'SBOM Format: {sbom.get("bomFormat", "CycloneDX")}')
print(f'Components: {len(sbom.get("components", []))}')
```

## 4. Incident Logging

```python
from neuralbridge.compliance.incident_log import IncidentLogger

incident_logger = IncidentLogger()
incident_id = incident_logger.log_incident(
    severity='low',
    title='Rate limit exceeded',
    description='Agent exceeded configured rate limit on Slack adapter.',
    affected_components=['slack_adapter'],
    reporter='system',
)
print(f'Incident logged: {incident_id}')
print(f'Total incidents: {len(incident_logger.list_incidents())}')
```

## 5. GDPR Article 30 Register

```python
from neuralbridge.compliance.gdpr_report import GDPRRegister, ProcessingActivity

register = GDPRRegister(controller_name='Your Company')
register.add_activity(ProcessingActivity(
    name='CRM Data Sync',
    purpose='Synchronise customer records via Salesforce adapter',
    data_categories=['name', 'email', 'phone'],
    legal_basis='Contract performance',
))

print(register.export_markdown())
```
