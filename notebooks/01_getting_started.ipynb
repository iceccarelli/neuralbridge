**NeuralBridge** is the missing infrastructure layer for agentic AI — an open-source, standardized adapter platform that lets ANY AI agent securely connect to ANY data source, API, or enterprise system.

## 1. Import and Create the Registry

The `AdapterRegistry` is the central hub that manages all your adapter connections.

```python
from neuralbridge.core.router import AdapterRegistry, RequestRouter
from neuralbridge.security.audit import AuditLogger

registry = AdapterRegistry()
audit = AuditLogger()
router = RequestRouter(registry=registry, audit_logger=audit)

print('Registry created successfully!')
```

## 2. Register Adapters

NeuralBridge supports 22+ adapters across databases, APIs, messaging, productivity, ERP/CRM, and cloud platforms.

```python
from neuralbridge.adapters.databases.postgres import PostgresAdapter
from neuralbridge.adapters.messaging.slack import SlackAdapter
from neuralbridge.adapters.erp_crm.salesforce import SalesforceAdapter

registry.register('postgres', PostgresAdapter(config={'host': 'localhost'}))
registry.register('slack', SlackAdapter(config={'bot_token': 'xoxb-demo'}))
registry.register('salesforce', SalesforceAdapter(config={}))

print(f'Registered adapters: {registry.list_all()}')
```

## 3. Execute Operations

Use the `RequestRouter` to execute operations on any registered adapter. All operations are automatically audited.

```python
import asyncio

async def run_queries():
    # Query PostgreSQL
    result = await router.route(
        adapter_type='postgres',
        operation='query',
        params={'sql': 'SELECT * FROM users LIMIT 5'}
    )
    print('PostgreSQL result:', result)
    
    # Query Salesforce
    result = await router.route(
        adapter_type='salesforce',
        operation='query',
        params={'soql': 'SELECT Id, Name FROM Account LIMIT 3'}
    )
    print('Salesforce result:', result)

await run_queries()
```

## 4. Review the Audit Trail

Every operation is logged in an immutable, hash-chain-verified audit trail — a core requirement for EU CRA compliance.

```python
events = await audit.query_events(filters={})
print(f'Total audit events: {len(events)}')
for event in events:
    print(f'  - {event}')
```

## 5. Generate a CRA Compliance Report

NeuralBridge can generate EU Cyber Resilience Act compliance reports on demand.

```python
from neuralbridge.compliance.cra_report import CRAReportGenerator

generator = CRAReportGenerator()
report = generator.generate_vulnerability_report()
print('CRA Report generated:')
print(report)
```
